```python
urls=["https://www.metro.pe/","https://www.plazavea.com.pe/","https://www.tottus.com.pe/"] 
```

```python
supermercados=["metro","plazavea","tottus"]

for i in supermercados: #juntar las palabras claves y el link de cada supermercado
  if i=="metro":
    urlb_metro=urls[0]+"busca/?ft="+producto.replace(" ", "%20")
  if i=="plazavea":
     urlb_plazavea=urls[1]+"search/?_query="+producto.replace(" ", "%20")
  if i=="tottus":
    urlb_tottus=urls[2]+"buscar?q="+producto.replace(" ", "%20")
```

## METRO

```python
req_metro = requests.get(urlb_metro) # Acceder a la página (recurso)
soup_metro = BeautifulSoup(req_metro.text, 'html.parser') # se tiene los datos anidados
name_metro = soup_metro.find("div", class_="product-shelf").find_all('a', class_ ='product-item__name') # El nombre del producto con su etiqueda
precios_metro=soup_metro.find("div", class_="product-shelf").find_all('span', class_ ='product-prices__value product-prices__value--best-price')
# Precio del producto
nombres_produc_metro=[]
precios_produc_metro=[]
```

```python
import pandas as pd 
for i in range(len(name_metro)): # guardar todos los productos (con la palabra clave) en su respectivo lista
  nombres_produc_metro.append(name_metro[i].text)
  precios_produc_metro.append(precios_metro[i].text)

    #Quitar los signos (s/, u) y cambiarlo a una variable numerica
precios_metroformato= [ float(i[4:].replace(",","")) for i in precios_produc_metro] 
precios_metroformato
mercado=["metro"]*len(precios_metroformato)
```

```python
# Guardar como dataframe
df_metro=pd.DataFrame({"Nombre_producto":nombres_produc_metro,"Precios_soles":precios_metroformato, "Supermercado":mercado })
```

```python
# Mostrar todo lo que se encuentra en este tag (class)
if soup_metro.find("div",class_="subtitle")==None:
  print(df_metro)
else:
  print("No se econtro resultados para la busqueda")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

options = webdriver.ChromeOptions() #Tener acceso al chrome
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=options) #chromedriver que está descargado
driver.implicitly_wait(10) # es como time.sleep pero para webdriver
driver.get(urlb_metro)
```

```python
# Para no tener una lista de listas
def flatten_list(_2d_list):
    flat_list = []
    # Iterar a través de la lista exterior
    for element in _2d_list:
        if type(element) is list:
            # Si el elemento es de tipo lista, iterar a través de la sublista
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
```

```python
import re

previous_height = driver.execute_script('return document.body.scrollHeight') #Automaticamente de ejecuta con javascript (java puede hacer lo que quiera en una web)

while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight);') #para que movilice hasta la última parte de la página
    time.sleep(3)
    
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == previous_height: #hasta que se cumnpla, se va almacenando todos los productos que va apareciendo en la web
        
        g_METRO = driver.find_elements(By.XPATH,".//a[@class='product-item__name']")
        nombre_METRO=[]
        for elemento in g_METRO:
            nombre_METRO.append(elemento.text)
            
        g_METRO_PRECIOS = driver.find_elements(By.XPATH,".//span[@class='product-prices__value product-prices__value--best-price']")
        precios_METRO=[]
        for elemento in g_METRO_PRECIOS:
            sentence = elemento.text
            s=re.findall(r'-?\d+\.?\d*', sentence)
            precios_METRO.append(s)
            precios_METRO=flatten_list(precios_METRO)
        
        break
    previous_height = new_height
    
   

#precios_METRO
#nombre_METRO
```

```python
#Crear una función para que tener varias tablas en un solo output
from IPython.display import display_html
from itertools import chain,cycle
def display_side_by_side(*args,titles=cycle([''])):
    html_str=''
    for df,title in zip(args, chain(titles,cycle(['</br>'])) ):
        html_str+='<th style="text-align:center"><td style="vertical-align:top">'
        html_str+=f'<h2 style="text-align: center;">{title}</h2>'
        html_str+=df.to_html().replace('table','table style="display:inline"')
        html_str+='</td></th>'
    display_html(html_str,raw=True)
```

```python
#datos estadísticos por cada supermercado
Descripción_metro = df_metro.describe() 
Descripción_plazavea = df_plazavea.describe()
Descripción_tottus = df_tottus.describe()

#aplicar la función creada (se muestra las 3 tablas)
display_side_by_side(Descripción_metro,Descripción_plazavea,Descripción_tottus, titles=['Metro','Plazavea','tottus'])
```

## Enviar al correo

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
 
# Iniciamos los parámetros del script
remitente = 'lp2.trabajofinal@gmail.com'
destinatarios = correo 
asunto = "Recopilación y comparación del producto solicitado"
cuerpo = 'Le adjunto el reporte con el producto solicitado'
ruta_adjunto = 'Reporte.xlsx'
nombre_adjunto = 'Reporte.xlsx'

# Creamos el objeto mensaje
mensaje = MIMEMultipart()
 
# Establecemos los atributos del mensaje
mensaje['From'] = remitente
mensaje['To'] = ", ".join(destinatarios)
mensaje['Subject'] = asunto
 
# Agregamos el cuerpo del mensaje como objeto MIME de tipo texto
mensaje.attach(MIMEText(cuerpo, 'plain'))
 
# Abrimos el archivo que vamos a adjuntar
archivo_adjunto = open(ruta_adjunto, 'rb')
 
# Creamos un objeto MIME base
adjunto_MIME = MIMEBase('application', 'octet-stream')
# Y le cargamos el archivo adjunto
adjunto_MIME.set_payload((archivo_adjunto).read())
# Codificamos el objeto en BASE64
encoders.encode_base64(adjunto_MIME)
# Agregamos una cabecera al objeto
adjunto_MIME.add_header('Content-Disposition', "attachment; filename= %s" % nombre_adjunto)
# Y finalmente lo agregamos al mensaje
mensaje.attach(adjunto_MIME)
 
# Creamos la conexión con el servidor
sesion_smtp = smtplib.SMTP('smtp.gmail.com', 587)
 
# Ciframos la conexión
sesion_smtp.starttls()

# Iniciamos sesión en el servidor
sesion_smtp.login('lp2.trabajofinal@gmail.com','ddwdfrhcxmoefajm')

# Convertimos el objeto mensaje a texto
texto = mensaje.as_string()

# Enviamos el mensaje
sesion_smtp.sendmail(remitente, destinatarios, texto)

# Cerramos la conexión
sesion_smtp.quit()
```
